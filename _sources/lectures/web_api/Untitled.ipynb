```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(6,4))
# Set the seed for reproducibility
np.random.seed(123)

# Generating some hypothetical data
true_values = np.random.normal(100, 15, 100)  # True number of grapefruits bought
predicted_values = true_values + np.random.normal(3, 5, 100)  # Predicted number, intentionally higher

# Creating the scatterplot
plt.figure(figsize=(8, 6), dpi=80)
plt.scatter(true_values, predicted_values, color='green')
plt.plot([80, 120], [80, 120], color='red', linestyle='--')  # Line y = x for reference

# Adding titles and labels
plt.title("Scatterplot of predicted vs. true number of kiwis bought")
plt.xlabel("True number of kiwis bought")
plt.ylabel("Predicted number of kiwis bought")
plt.grid(True)
# Show the plot
plt.show()
```
